```python
import pandas as pd
import numpy as np
import json
import requests
```

```python
base = 'https://swapi.dev/api/people/'
people = requests.get(base).json()
```

```python
people['count']
```

```python
results = people['results'] # a list of dictionary

while people['next']:
    people = requests.get(people['next']).json()
    results = results + people['results']
```

```python
len(results)
```

```python
people_df = pd.DataFrame(results)
```

```python
people_df.shape
```

```python
people_df.head()
```

```python
BBY = people_df[people_df.birth_year.str.endswith('BBY')]
BBY = BBY.assign(birth_year = BBY['birth_year'].map(
    lambda x: float(x.rstrip('BBY'))))
BBY.head()
```
