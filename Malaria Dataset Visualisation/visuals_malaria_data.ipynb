## Malaria Dataset

__3 Datasets:__
* [`malaria_inc.csv`](https://github.com/rfordatascience/tidytuesday/blob/master/data/2018/2018-11-13/malaria_inc.csv) - Malaria incidence by country for all ages across the world across time
* [`malaria_deaths.csv`](https://github.com/rfordatascience/tidytuesday/blob/master/data/2018/2018-11-13/malaria_deaths.csv) - Malaria deaths by country for all ages across the world and time.
* [`malaria_deaths_age.csv`](https://github.com/rfordatascience/tidytuesday/blob/master/data/2018/2018-11-13/malaria_deaths_age.csv) - Malaria deaths by age across the world and time.



### Three informative visualizations about Malaria

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# load in datasets
inc = pd.read_csv('data/malaria_inc.csv')
death = pd.read_csv('data/malaria_deaths.csv')
age = pd.read_csv('data/malaria_deaths_age.csv')
```

### plot 1

```python
# restructure and extract information

inc['Incidence'] = (inc.iloc[:, 3])
#grouped.groups.keys()
grouped = inc.groupby('Entity')
df1 = grouped.get_group('Low income') # eg, 'Central African Republic'
df2 = grouped.get_group('Lower middle income') # eg, 'Nepal'
df3 = grouped.get_group('Middle income') # eg, 'Bangladesh'
df4 = grouped.get_group('Upper middle income') # eg, 'China'
newdf1 = pd.concat([df1, df2, df3, df4])
```

```python
newdf1
```

```python
g = sns.catplot(x="Year", y="Incidence", col="Entity",
                data=newdf1, saturation=.5,
                kind="bar", ci=None, aspect=.5)
(g.set_axis_labels("Year", "Incidence of malaria (per 1,000 population at risk)")
  .set_xticklabels(["2000", "2005", "2010", "2015"])
  .set_titles("{col_name} {col_var}")
  .despine(left=True)
  .fig.suptitle('Incidence of Malaria over Population at Risk for Different Income Entities (2000 - 2015)',
                y=1.05))
g.savefig('plot1.png')
pass
```

### plot 2

```python
# restructure and extract information

death['Deaths'] = (death.iloc[:, 3])
death_new = death[(death.Year == 2000) | (death.Year == 2005) 
                  | (death.Year == 2010) | (death.Year == 2015)]
#grouped.groups.keys()
grouped = death_new.groupby('Entity')
df1 = grouped.get_group('Central African Republic')
df2 = grouped.get_group('Nepal')
df3 = grouped.get_group('Bangladesh')
df4 = grouped.get_group('China')
newdf2 = pd.concat([df1, df2, df3, df4])
```

```python
newdf2
```

```python
g = sns.catplot(x="Year", y='Deaths', col="Entity",
                data=newdf2, saturation=.5,
                kind="bar", ci=None, aspect=.5)
(g.set_axis_labels("Year", "Deaths of malaria (per 100,000 popolation)")
  .set_xticklabels(["2000", "2005", "2010", "2015"])
  .set_titles("{col_name}")
  .despine(left=True)
  .fig.suptitle('Deaths of Malaria over Whole Population for Four Example-different-income Countries (2000 - 2015)',
                y=1.05))
g.savefig('plot2.png')
pass
```

### plot 3

```python
# restructure and extract information

age_new = age.groupby('entity').get_group('Central African Republic')
newdf3 = age_new[(age_new.year == 2000) | (age_new.year == 2005)
                 | (age_new.year == 2010) | (age_new.year == 2015)]
```

```python
newdf3
```

```python
g = sns.catplot(x="year", y='deaths', col="age_group",
                data=newdf3, saturation=.5,
                kind="bar", ci=None, aspect=.4)
(g.set_axis_labels("Year", "Deaths of malaria")
  .set_xticklabels(["2000", "2005", "2010", "2015"])
  .set_titles("{col_name} {col_var}")
  .despine(left=True)
  .fig.suptitle('Deaths of Malaria for Different Age Groups in Central African Republic (2000 - 2015)',
                y=1.05))
g.savefig('plot3.png')
pass
```
